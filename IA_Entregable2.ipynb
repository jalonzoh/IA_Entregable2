## **Revision de las columnas inpactantes para el analisis**

```python
#Verificando Valores unicos en cada una de las Columnas
for col in bdAurelion.columns:
  print('-' * 60 )
  print(f"Valores unicos en la columna {col}: {bdAurelion[col].nunique()}")
  if bdAurelion[col].nunique() < 10:
    print(f"Valores: {list(bdAurelion[col].unique())}")
    print('=' * 60)
```

```python
#Validando Valores Duplicados.
duplicados=bdAurelion.duplicated().sum()
print(f"Cantidad de lineas duplicados: {duplicados}")
```

```python
#Validando Valores Nulos
valores_nulos=bdAurelion.isnull().values.sum()
print("Cantidad de valores nulos por columna:")
print(valores_nulos[valores_nulos >0 ])
```

```python
#Validando valores en blanco
valores_en_blanco=bdAurelion.apply(lambda x: x.astype(str).str.strip()=='').sum()
print("\nCantidad de valores en blanco o string por columna:")
print(valores_en_blanco[valores_en_blanco >0 ])
```

```python
#Cambiando columnas que que deben ser tipo string
for columna in bdAurelion.columns:
    if bdAurelion[columna].dtype == 'object':
        try:
            # CorrecciÃ³n: usar 'columna' en singular
            bdAurelion[columna] = bdAurelion[columna].astype(pd.StringDtype())
            # Opcional: imprimir si la conversiÃ³n fue exitosa para esa columna
            # print(f"Columna '{columna}' convertida a str.")
        except Exception as e:
            print(f'No pudo convertirse la columna {columna} a texto')
```

```python
# Eliminando las columnas vacias de Fecha_alta
bdAurelion = bdAurelion[bdAurelion['fecha_alta'].str.strip() != '']
print("NÃºmero de filas despuÃ©s de eliminar las vacÃ­as en 'Churn':", len(bdAurelion))
```

```python
# Despues de realizar las validaciones se vuelve a comprobar si aun hay nulos o valores en blanco
# VacÃ­os o en blanco
bdAurelion.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
# Una vez realizado el ETL podemos realizamos el analisis

import plotly.express as px
```

```python
#Cantidad total de clientes por categoria de producto
fig=px.histogram(bdAurelion, x = 'Categoria', text_auto = True, color = 'Categoria', barmode = 'group')
fig.show()

#Porcentaje total de clientes por medio de pago
porcentaje_churn = round(bdAurelion['medio_pago'].value_counts(normalize=True)*100, 2)
names = ['efectivo', 'qr', 'tarjeta','transferencia']
fig=px.pie(bdAurelion,values=porcentaje_churn, names=names, title='Porcentaje por medios de pago')
fig.show()
```

## **Graficos para el analisis de la Columna Churn (Se fue de la empresa o no) y las variables categoricas**

```python
columnas_categoricas = ['ciudad', 'medio_pago']

numero_columnas = 2
numero_filas = (len(columnas_categoricas) + numero_columnas - 1) // numero_columnas # Ensure enough rows for all plots
fig, ax = plt.subplots(numero_filas, numero_columnas, figsize=(numero_columnas*7, numero_filas*5))
ax = ax.flatten()
plt.suptitle('AnÃ¡lisis de Categorias de productos por Variables CategÃ³ricas relevantes', fontsize=16, y=1.02, fontweight = 'bold')

for i, columna in enumerate(columnas_categoricas):
  if i < len(ax):

    sns.countplot(x=columna, hue='Categoria', data=bdAurelion, ax=ax[i], palette='Greens')
    ax[i].set_title(f'RelaciÃ³n categoria de producto y {columna}', fontsize = 12)
    ax[i].set_xlabel(columna, fontsize = 11)
    ax[i].set_ylabel('Cantidad de clientes')
    ax[i].tick_params(axis='x', rotation=45)
    ax[i].legend(title='Categoria', labels=['Limpieza', 'Alimentos'])
    ax[i].grid(axis='y', linestyle='-', alpha=0.8)
    ax[i].bar_label(ax[i].containers[0])
    ax[i].bar_label(ax[i].containers[1])

# Ocultar los subplots vacÃ­os si los hay
for j in range(i + 1, len(ax)):
    fig.delaxes(ax[j])

plt.subplots_adjust(top=1.2, right=0.8)
plt.tight_layout()
plt.show()
```

```python
# Agrupar por cantidad y categoria
df_grouped = bdAurelion.groupby(['cantidad', 'Categoria']).size().unstack(fill_value=0)

# Calcular el % de cantidad de comprar productos categoria Limpieza
df_grouped['categoria_rate1'] = round((df_grouped.get('Limpieza', 0) / df_grouped.sum(axis=1)) * 100,1)

# Calcular el % de cantidad de comprar productos categoria Alimentos
df_grouped['categoria_rate2'] = round((df_grouped.get('Alimentos', 0) / df_grouped.sum(axis=1)) * 100,1)
```

```python
fig = px.line(df_grouped.reset_index(),
              x='cantidad',
              y='categoria_rate1',
              markers=True,
              title='Porcentaje de compra categoria Limpieza')
fig.show()

fig = px.line(df_grouped.reset_index(),
              x='cantidad',
              y='categoria_rate2',
              markers=True,
              title='Porcentaje de compra categoria Alimentos')
fig.show()
```

```python
fig = px.scatter(bdAurelion, x="cantidad", y="precio_unitario", size="importe", color="Categoria",
           hover_name="ciudad", log_x=False, size_max=10,title="Cantidad Vendida - Precio Unitario - Importe Total" )
fig.show()
```
